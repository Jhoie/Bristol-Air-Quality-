## DATA CLEANING

```python
# check the data types of the various columns
df.dtypes
```

Date Time, DateStart, and DateEnd columns are datetime columns, but they are recognized as objects by pandas
SiteID is an integer column, but it is recognized as a float by pandas.

Our data types need to be in the right format if we are to carry out any meaningful work on our data. 

```python
# convert Date Time, DateStart, and DateEnd columns from object to datetime
df["Date Time"]= pd.to_datetime(df["Date Time"], utc=True)
df["DateStart"]= pd.to_datetime(df["DateStart"], utc=True)
df["DateEnd"]= pd.to_datetime(df["DateEnd"], utc=True)
```

```python
# notice the change
df.dtypes
```

```python
# let's do the same for SiteID
df = df.astype({"SiteID":'Int64'})
```

```python
# notice the change
df.dtypes
```

```python
# PLEASE NOTE: The changes made in the dataframe do not affect the original dataset. It only makes pandas understand the datat types in the columns 
df
```
